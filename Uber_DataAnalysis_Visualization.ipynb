# Data Cleaning and Wrangling

```
df_2014 = pd.concat([april, may, june, july, august, september])
```

```
df_2014
```

```
df_2014[["Date", "Time"]] = df_2014["Date/Time"].str.split(expand = True,)
df_2014.drop(["Date/Time"], axis=1, inplace=True)
df_2014
```

```
df_2014["Base"].replace({"B02512":"Unter","B02598":"Hinter","B02617":"Weiter","B02682":"Schmecken",
                    "B02764":"Danach-NY","B02765":"Grun","B02835":"Dreist","B02836":"Drinnen"}, inplace=True)
df_2014
```

```
x = df_2014["Time"].replace({r'^0.*$': "0:00",
                         r'^1.*$': "1:00",
                         r'^2.*$': "2:00",
                         r'^3.*$': "3:00",
                         r'^4.*$': "4:00",
                         r'^5.*$': "5:00",
                         r'^6.*$': "6:00",
                         r'^7.*$': "7:00",
                         r'^8.*$': "8:00",
                         r'^9.*$': "9:00",
                         r'^1[0].*$': "10:00",
                         r'^1[1].*$': "11:00",
                         r'^1[2].*$': "12:00",
                         r'^1[3].*$': "13:00",
                         r'^1[4].*$': "14:00",
                         r'^1[5].*$': "15:00",
                         r'^1[6].*$': "16:00",
                         r'^1[7].*$': "17:00",
                         r'^1[8].*$': "18:00",
                         r'^1[9].*$': "19:00",
                         r'^2[0].*$': "20:00",
                         r'^2[1].*$': "21:00",
                         r'^2[2].*$': "22:00",
                         r'^2[3].*$': "23:00"}, regex=True, 
                        inplace=True)

x 
```
